### Task 1: Trends in Compensation Over Age Groups


```python

!pip install seaborn
import seaborn as sns
```

##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# Convert the 'Age' column to different age groups for simplicity and better understanding
age_mapping = {
    'Under 18 years old': 'Under 18', '18-24 years old': '18-24', '25-34 years old': '25-34',
    '35-44 years old': '35-44', '45-54 years old': '45-54', '55-64 years old': '55-64',
    '65 years or older': '65+', 'Prefer not to say': 'Prefer not to say'
}
df['AgeGroups'] = df['Age'].map(age_mapping)

# Calculate the median yearly compensation (ConvertedCompYearly) of each age group
median_comp_by_age = df.groupby('AgeGroups')['ConvertedCompYearly'].median().reset_index()

# Plot the line chart of Median ConvertedCompYearly by Age Group
plt.figure(figsize=(10, 8))
sns.lineplot(data=median_comp_by_age,x='AgeGroups', y='ConvertedCompYearly',marker='o',linestyle='-',color='r')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.title('Median Yearly Compensation by Age Group')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
#Use the previously converted Agegroups data 
# Filter data for ages between 25 and 45
df_age_25_44 = df[df['AgeGroups'].isin(['25-34', '35-44'])]
# Calculate the median yearly compensation (ConvertedCompYearly) of each age group
median_comp_by_age_25_44 = df_age_25_44.groupby('AgeGroups')['ConvertedCompYearly'].median().reset_index()

# Plot the line chart of Median ConvertedCompYearly for Ages 25 to 45
plt.figure(figsize=(10, 8))
sns.lineplot(median_comp_by_age_25_44,x='AgeGroups', y='ConvertedCompYearly',marker='o',linestyle='-',color='r')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.title('Median Yearly Compensation for Ages 25 to 45')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# substitute an available experience-related column for Experience - 'WorkExp'
df=df.dropna(subset=['WorkExp'],axis=0)
jobsat_exp = df.groupby('WorkExp')['JobSatPoints_6'].median().reset_index()

# Plot the line chart of Job Satisfaction (JobSatPoints_6) by Experience Level
plt.figure(figsize=(10, 8))
sns.lineplot(jobsat_exp,x='WorkExp', y='JobSatPoints_6',marker='o',linestyle='-',color='r')
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
# Calculate Median ConvertedCompYearly Over Experience Level
median_comp_by_exp = df.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index()

# Plot the Line Chart of Median ConvertedCompYearly Over Experience Level
plt.figure(figsize=(10, 8))
sns.lineplot(median_comp_by_exp,x='WorkExp', y='ConvertedCompYearly',marker='o',linestyle='-',color='r')
plt.title('Median Compensation over Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Median Yearly Compensation')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
jobsat_exp = df.groupby('WorkExp')['JobSatPoints_6'].median().reset_index()

# Plot the line chart of Job Satisfaction (JobSatPoints_6) by Experience Level
plt.figure(figsize=(10, 8))
sns.lineplot(jobsat_exp,x='WorkExp', y='JobSatPoints_6',marker='o',linestyle='-',color='r')
plt.title('Job Satisfaction by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
