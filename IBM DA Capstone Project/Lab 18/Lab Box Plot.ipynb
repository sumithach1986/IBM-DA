## Visualizing Data


### Task 1: Visualizing the Distribution of Data


**1. Box Plot of `CompTotal` (Total Compensation)**


Use a box plot to analyze the distribution and outliers in total compensation.


```python
# your code goes here
QUERY = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL AND CompTotal < 1000000;"
df = pd.read_sql_query(QUERY, conn)

plt.boxplot(df['CompTotal'])
plt.title('Distribution of Total Compensation')
plt.xlabel('Total Compensation (CompTotal)')
plt.ylabel('Frequency')
plt.show()
```

**2. Box Plot of Age (converted to numeric values)**


Convert the `Age` column into numerical values and visualize the distribution.


```python
import seaborn as sns
```

```python
query = """
SELECT Age,
    CASE 
        WHEN Age = 'Under 18 years old' THEN 15
        WHEN Age = '18-24 years old' THEN 21
        WHEN Age = '25-34 years old' THEN 30
        WHEN Age = '35-44 years old' THEN 40
        WHEN Age = '45-54 years old' THEN 50
        WHEN Age = '55-64 years old' THEN 60
        WHEN Age = '65 years or older' THEN 70
        ELSE NULL
    END AS AgeNumeric
FROM main
WHERE Age IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

# Plot
plt.figure(figsize=(10,8))
sns.boxplot(data=df, y='AgeNumeric')
plt.title('Box Plot of Age (Converted to Numeric Values)')
plt.ylabel('Age (Numeric Values)')
plt.show()
```

### Task 2: Visualizing Relationships in Data


**1. Box Plot of `CompTotal` Grouped by Age Groups:**


Visualize the distribution of compensation across different age groups.


```python
#your code goes here
query = """
SELECT CompTotal, Age 
FROM main 
WHERE CompTotal IS NOT NULL AND Age IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)


# Create Boxplot of CompTotal Grouped by Age Groups
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='Age', y='CompTotal')
plt.title('Box Plot of Compensation by Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Total Compensation')
plt.yscale('log') #Apply log scale 
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot - After filtering out extreme values
query = """
SELECT CompTotal, Age 
FROM main 
WHERE CompTotal IS NOT NULL AND Age IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')
df = df[df['CompTotal'] <= 1_000_000]

# Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Age', y='CompTotal')
plt.title('Box Plot of Compensation by Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Total Compensation')
plt.yscale('log')
plt.xticks(rotation=45)
plt.show()
```

**2. Box Plot of `CompTotal` Grouped by Job Satisfaction (`JobSatPoints_6`):**


Examine how compensation varies based on job satisfaction levels.


```python
# your code goes here
query = """
SELECT CompTotal, JobSatPoints_6
FROM main 
WHERE CompTotal IS NOT NULL AND JobSatPoints_6 IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

#Create Box Plot
plt.figure(figsize=(14, 10))
ax=sns.boxplot(data=df, x='JobSatPoints_6', y='CompTotal')
plt.title('Box Plot of Compensation by Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Total Compensation')
plt.xticks(rotation=45)
xticks=ax.xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i%2==1:
        xticks[i].set_visible(False)
plt.show()
```

### Task 3: Visualizing the Composition of Data


**1. Box Plot of `ConvertedCompYearly` for the Top 5 Developer Types:**


Analyze compensation across the top 5 developer roles.


```python
# your code goes here
query = "SELECT DevType, ConvertedCompYearly FROM main WHERE ConvertedCompYearly IS NOT NULL AND DevType IS NOT NULL"
df = pd.read_sql_query(query, conn)

# Get the top 5 developer types
top_5_devtypes = df['DevType'].value_counts().nlargest(5).index
df_top_5 = df[df['DevType'].isin(top_5_devtypes)]

# Plot the box plot for ConvertedCompYearly grouped by DevType
plt.figure(figsize=(14, 8))
sns.boxplot(x='DevType', y='ConvertedCompYearly', data=df_top_5)
plt.title('Distribution of Total Compensation for Top 5 Developer Roles')
plt.xlabel('Developer Role')
plt.ylabel('Total Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

**2. Box Plot of `CompTotal` for the Top 5 Countries:**


Analyze compensation across respondents from the top 5 countries.


```python
# your code goes here
query = """
SELECT CompTotal, Country 
FROM main 
WHERE CompTotal IS NOT NULL AND Country IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

# Get the top 5 countries
top_5_countries = df['Country'].value_counts().nlargest(5).index
df_top5 = df[df['Country'].isin(top_5_countries)]

# Plot the box plot of CompTotal for the Top 5 Countries
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_top5,x='Country', y='CompTotal')
plt.title('Distribution of Total Compensation for Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Total Compensation')
plt.xticks(rotation=45)
plt.yscale('log')
plt.show()
```

### Task 4: Visualizing Comparison of Data


**1. Box Plot of CompTotal Across Employment Types:**


Analyze compensation for different employment types.


```python
# your code goes here
query = """
SELECT CompTotal, Employment 
FROM main 
WHERE CompTotal IS NOT NULL AND Employment IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

# Plot
plt.figure(figsize=(14, 10))
ax=sns.boxplot(data=df, x='Employment', y='CompTotal')
plt.title('Box Plot of Compensation across Employment Types')
plt.xlabel('Employment Types')
plt.ylabel('Total Compensation')
plt.yscale('log')
plt.xticks(rotation=45)
xticks=ax.xaxis.get_major_ticks()
for i in range(len(xticks)):
    if i%2==1:
        xticks[i].set_visible(False)
plt.show()
```

```python
# Simplified Box Plot after mapping different types of employment into simpler forms
query = "SELECT Employment, CompTotal FROM main WHERE CompTotal IS NOT NULL AND Employment IS NOT NULL"
df = pd.read_sql_query(query, conn)

# Simplify the employment types for readability
employment_mapping = {
    'Employed, full-time': 'Full-time',
    'Employed, part-time': 'Part-time',
    'Independent contractor, freelancer, or self-employed': 'Self-employed',
    'Not employed, and not looking for work': 'Not looking',
    'Not employed, but looking for work': 'Looking for work',
    'Retired': 'Retired',
    'Student, full-time': 'Full-time student',
    'Student, part-time': 'Part-time student',
    'Other': 'Other'
}
df['Employment']=df['Employment'].dropna()
df['CompTotal']=df['CompTotal'].dropna()
df['Simplified_Employment'] = df['Employment'].map(employment_mapping)

# Create boxplot
plt.figure(figsize=(14, 8))
sns.boxplot(x='Simplified_Employment', y='CompTotal', data=df)
plt.title('Box Plot of Compensation across Employment Types')
plt.xlabel('Employment Types')
plt.ylabel('Total Compensation')
plt.yscale('log')
plt.show()
```

**2. Box Plot of `YearsCodePro` by Job Satisfaction (`JobSatPoints_6`):**


Examine the distribution of professional coding years by job satisfaction levels.


```python
# your code goes here
query = """
SELECT YearsCodePro, JobSatPoints_6 
FROM main 
WHERE YearsCodePro IS NOT NULL AND JobSatPoints_6 IS NOT NULL;
"""
df = pd.read_sql_query(query, conn)

df['YearsCodePro']=df['YearsCodePro'].dropna()
df['JobSatPoints_6']=df['JobSatPoints_6'].dropna()

# Plot
plt.figure(figsize=(12, 14))
sns.boxplot(x='JobSatPoints_6', y='YearsCodePro', data=df)
plt.title('Box Plot of Years of Coding Experience by Job Satisfaction Levels')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Years of Professional Coding Experience')
plt.show()
```
